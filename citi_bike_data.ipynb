```python
import pandas as pd
import datetime as dt
```

```python
citiBike_data01 = pd.read_csv("Data/202101-citibike-tripdata.csv")
citiBike_data02 = pd.read_csv("Data/202102-citibike-tripdata.csv")
citiBike_data03 = pd.read_csv("Data/202103-citibike-tripdata.csv")
citiBike_data04 = pd.read_csv("Data/202104-citibike-tripdata.csv")

citiBike_data01 = pd.DataFrame(citiBike_data01)
citiBike_data02 = pd.DataFrame(citiBike_data02)
citiBike_data03 = pd.DataFrame(citiBike_data03)
citiBike_data04 = pd.DataFrame(citiBike_data04)
```

```python
# Concatenate csv files into one file
citiBike_data = pd.concat([citiBike_data01, citiBike_data02, citiBike_data03, citiBike_data04])

citiBike_data
```

```python
# Convert duration column from seconds to minutes
citiBike_data['tripduration'] = citiBike_data['tripduration'].astype('float') 
citiBike_data['tripduration'] = (citiBike_data['tripduration'] / 60).round(2)
# Dropping outlier data
citiBike_data = citiBike_data[citiBike_data['tripduration'] < 100]

citiBike_data
```

```python
# Convert gender column to words
citiBike_data['gender'] = citiBike_data['gender'].astype('str') 
citiBike_data['gender'].replace({"1": "male", "2": "female", "0": "unknown"}, inplace=True)

citiBike_data
```

```python
# Rename Columns
citiBike_data = citiBike_data.rename(columns={'tripduration': 'Trip Duration (Minutes)',
                                              'starttime': 'Start Time',
                                              'stoptime': 'Stop Time',
                                              'start station id': 'Start Station ID',
                                              'start station name': 'Start Station Name',
                                              'start station latitude': 'Start Station Latitude',
                                              'start station longitude': 'Start Station Longitude', 
                                              'end station id': 'End Station ID',
                                              'end station name': 'End Station Name',
                                              'end station latitude': 'End Station Latitude',
                                              'end station longitude': 'End Station Longitude',
                                              'bikeid': 'Bike ID',
                                              'usertype': 'User Type',
                                              'birth year': 'Birth Year',
                                              'gender': 'Gender'})
citiBike_data
```

```python
# Export file
citiBike_data.to_csv("Data/citiData.csv")
```
